##**Creating Plots 101**
In Matplotlib, there are three most important terms.
1. A figure is the container for all elements of a plot
2. Axes (subplots) is the area that contains the plot. It is the region of the figure where data is visualized. It represents an individual plot or subplot within the figure.A figure can have one or more Axes.Each Axes has its own x- and y-scales.
3. A plot is the actual representation of data on an Axes.It can be a line plot, scatter plot, bar plot, etc.

Figures are created with plt.figure()
Axes are created with plt.subplots() or plt.subplot()
The simplest way of creating a Figure with an Axes is using pyplot.subplots()

##**Creating Plots 102**
Matplotlib supports a diverse range of plot types, including:

**Line Plots:** Display data points connected by straight lines, useful for showing trends over a continuous variable.

Scatter Plots:Represent individual data points on a two-dimensional plane, helpful for visualizing relationships and patterns.

**Bar Plots:**Illustrate data using rectangular bars, commonly used for comparing values across different categories.

**Histograms:**Visualize the distribution of a dataset by grouping data into bins and plotting the frequency of occurrences.

**3D Plots:**Generate three-dimensional visualizations, such as scatter plots or surface plots, to represent data in a spatial context.

**Contour Plots:**Display three-dimensional data on a two-dimensional plane using contour lines, useful for visualizing data with varying levels.

**Heatmaps:** Visualize data in a matrix format with colors indicating the intensity of values, often used for correlation matrices or data with two categorical variables.

**Pie Charts:** Represent data as slices of a circular pie, useful for showing the proportion of each category in a dataset.

**Box Plots:**Display the distribution of a dataset and highlight statistical measures such as median, quartiles, and outliers.

##**Customizing Your Figure's Appearance and Layout**

When creating a figure in Matplotlib, you have several options to tailor its appearance and layout:

#**A. Size and Resolution**
1. **Figsize:** This controls the figure's physical dimensions on the screen. It's specified as a tuple of two values: (width in inches, height in inches).
2.** DPI (Dots Per Inch):** This determines the resolution of the figure when saved as an image. It specifies how many pixels will be used to represent each inch of the figure. Higher DPI means sharper images but larger file sizes.
3. **Look and Feel Facecolor:** This sets the background color of the figure. **Edgecolor:** This sets the color of the figure's border. Linewidth: This adjusts the thickness of the border. Frameon: If set to False, this removes the border entirely, making the figure transparent.

#**B.Layout**
Matplotlib offers different layout engines to automatically arrange elements within a figure and prevent overlapping labels.

1. "constrained" Optimizes layout while preserving aspect ratios.
2. "compressed" Tightly packs elements, potentially altering aspect ratios.
3. "tight"Combines both approaches for a balanced layout.

Customization options options to enhance the visual appeal include:
1. Labels: plt.xlabel(), plt.ylabel(), plt.title()

```python
# Create a figure with set y_label and x_label
# Generate random data with mean (mu) and standard deviation (sigma)
mu, sigma = 115, 15
x = mu + sigma * np.random.randn(10000)

# Create a figure and axes with a specified size and constrained layout
fig, ax = plt.subplots(figsize=(5, 2.7), constrained_layout=True)

# Plot a histogram of the data
n, bins, patches = ax.hist(x, 50, density=True, facecolor='g', alpha=0.75)

# Set labels, title, and axis limits
ax.set_xlabel('Length [cm]')
ax.set_ylabel('Probability')
ax.set_title('Practice')

# Add text annotation with mean and standard deviation
ax.text(75, 0.025, r'$\mu=115,\ \sigma=15$')

# Set axis limits
ax.axis([55, 175, 0, 0.03])

# Add grid lines
ax.grid(True)
```

```python
# Annotation
# Points on the plot can often be annotated by connecting
# an arrow pointing to xy,to a piece of text at xytext

# Create a figure and axes with a specified size
fig, ax = plt.subplots(figsize=(5, 2.7))

# Sample data
t = np.arange(0.0, 6.0, 0.02)
s = np.sin(2 * np.pi * t)

# Plot the sine curve
line, = ax.plot(t, s, lw=2)

# Annotate a point on the curve
ax.annotate('local max', xy=(3, 1), xytext=(4, 1.5),
            arrowprops=dict(facecolor='black', shrink=0.05),
            fontsize=10, color='red', style='italic',
            bbox=dict(boxstyle='round', alpha=0.2))

# Set y-axis limits
ax.set_ylim(-1.5, 1.5)
```

```python
# Another anotation example
# Create a figure and axes with a specified size
fig, ax = plt.subplots(figsize=(3, 3))

# Sample data
x_values = [2, 4, 6, 8, 10]
y_values = [1, 3, 5, 7, 9]
annotations = ["Point A", "Point B", "Point C", "Point D", "Point E"]

# Scatter plot
ax.scatter(x_values, y_values, s=20, color='green')

# Annotate each point with text
for xi, yi, text in zip(x_values, y_values, annotations):
    ax.annotate(text,
                xy=(xi, yi), xycoords='data',
                xytext=(5, 5), textcoords='offset points',
                arrowprops=dict(facecolor='black', arrowstyle='wedge,tail_width=0.7', lw=1),
                fontsize=8, color='blue', ha='right', va='bottom')
```

#**Axis scales and Ticks**
Axis are objects that represent the x-axis and y-axis
Each Axes has two or three Axis
Tick locators and tick formatters are used to control the scale of Axis objects

In this context "scales" typically refers to the mapping between data values and the positions or lengths along an axis. Scales help define how data points are represented on a plot and how they relate to each other in terms of spacing and positioning.

The common scales types are Linear and Logarithmic scales

```python
# Linear and log scales examples

# Sample data for linear scale
data2_linear = np.random.rand(10)

# Create a figure with two subplots (linear scale and log scale)
fig, axs = plt.subplots(1, 2, figsize=(8, 3), constrained_layout=True)

# Plot on the left subplot with linear scale
axs[0].plot(np.arange(len(data2_linear)), data2_linear)
axs[0].set_title('Linear Scale')

# Sample data for log scale
data2_log = np.random.randint(1, 100, size=10)

# Plot on the right subplot with log scale
axs[1].set_yscale('log')
axs[1].plot(np.arange(len(data2_log)), data2_log)
axs[1].set_title('Log Scale')
```

#**Plotting Dates and Strings**

Matplotlib has the capability to create plots using arrays of dates, arrays of strings, and floating-point numbers. Depending on the data type, Matplotlib automatically applies specific locators and formatters to enhance the representation.

```python
# Ploting Dates

# Create a figure and axes with a specified size and constrained layout
fig, ax = plt.subplots(figsize=(5, 2.7), constrained_layout=True)

# Generate sample date data and cumulative sum of random data
dates = np.arange(np.datetime64('2022-01-01'), np.datetime64('2022-01-15'),
                  np.timedelta64(1, 'h'))
data = np.cumsum(np.random.randn(len(dates)))

# Plot the data
ax.plot(dates, data)

# Use ConciseDateFormatter for the x-axis
cdf = mpl_dates.ConciseDateFormatter(ax.xaxis.get_major_locator())
ax.xaxis.set_major_formatter(cdf)
```

```python
# Plotting strings
# For strings, we get categorical plotting

# Create a figure and axes with a specified size and constrained layout
fig, ax = plt.subplots(figsize=(5, 2.7), constrained_layout=True)

# Sample data representing the number of casualties in different age categories
categories = ['Children', 'Teens', 'Adults', 'Elderly']
casualties = np.random.randint(1, 50, len(categories))

# Plot a bar chart for age categories
ax.bar(categories, casualties, color='red')

# Set labels and title
ax.set_xlabel('Age Categories')
ax.set_ylabel('Number of Casualties')
ax.set_title('Casualties in Gaza-Israel Conflict')
```

```python
# Created a number of ColorMaps
# Colors in a colormap are used when we want to have a third dimension in the plot represented by colors

# Create different data for the plot
X_new, Y_new = np.meshgrid(np.linspace(-2, 2, 128), np.linspace(-2, 2, 128))
Z_new = np.sin(X_new**2 + Y_new**2) * np.exp(-X_new**2 - Y_new**2)

data1 = np.random.rand(50)
data2 = np.random.rand(50)
data3 = np.random.rand(50)

# Create a figure with 2x2 subplots
fig, axs = plt.subplots(2, 2, constrained_layout=True)

# Plot 1: pcolormesh
pc1 = axs[0, 0].pcolormesh(X_new, Y_new, Z_new, vmin=-0.5, vmax=0.5, cmap='inferno')
fig.colorbar(pc1, ax=axs[0, 0])
axs[0, 0].set_title('pcolormesh()')

# Plot 2: contourf
co = axs[0, 1].contourf(X_new, Y_new, Z_new, levels=np.linspace(-0.7, 0.7, 11), cmap='coolwarm')
fig.colorbar(co, ax=axs[0, 1])
axs[0, 1].set_title('contourf()')

# Plot 3: imshow with LogNorm
pc2 = axs[1, 0].imshow(Z_new**2 * 100, cmap='rainbow',
                       norm=mpl.colors.LogNorm(vmin=0.01, vmax=100))
fig.colorbar(pc2, ax=axs[1, 0], extend='both')
axs[1, 0].set_title('imshow() with LogNorm()')

# Plot 4: scatter plot
sc = axs[1, 1].scatter(data1, data2, c=data3, cmap='plasma')
fig.colorbar(sc, ax=axs[1, 1], extend='both')
axs[1, 1].set_title('scatter()')
```

##**Interactive Figures**

Interactive figures are dynamic visualizations that allow users to engage with the plot elements in real-time, offering features such as zooming, panning, and additional information on hovering.

Using the Plotly library in Python, interactive figures can be easily created by defining traces, configuring layout options, and enabling interactive components.

```python
import plotly.graph_objects as go

# Create sample data
x_data = np.linspace(0, 10, 100)
y_data = np.sin(x_data)

# Create an interactive scatter plot
fig = go.Figure()

scatter = fig.add_trace(go.Scatter(x=x_data, y=y_data, mode='lines', name='Sine Wave'))

# Set layout options for interactivity
fig.update_layout(
    title='Interactive Sine Wave',
    xaxis_title='X-axis',
    yaxis_title='Y-axis',
    xaxis=dict(type='linear'),
    yaxis=dict(type='linear'),
    showlegend=True
)

# Enable zooming and panning
fig.update_layout(
    xaxis=dict(
        rangeslider=dict(
            visible=True
        ),
        type='linear'
    )
)

# Show the interactive plot in Colab
fig.show()
```

#**Helper Functions**
Helper functions are needed when you want to make the same plots over and over again using different data sets or when you want to easily wrap matplotlib methods.

When making the helper functions, it is recommended to use the signature function below:
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph.
    """
    out = ax.plot(data1, data2, **param_dict)
    return out


```python
# Plots made with Helper functions
def my_line_plotter(ax, x_data, y_data, param_dict):
    out = ax.plot(x_data, y_data, **param_dict)
    return out

def create_and_show_plot(ax, x_values, y_values, label, color, title):
    # Create a plot using the my_line_plotter function and display it on a specified Axes.
    my_line_plotter(ax, x_values, y_values, {'label': label, 'color': color})
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_title(title)
    ax.legend()

# Create a single figure with two subplots (side by side)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

# Dataset 1:
x_values_1 = [1, 2, 3, 4, 5]
y_values_1 = [10, 15, 7, 22, 12]
create_and_show_plot(ax1, x_values_1, y_values_1, 'My Line Plot', 'blue', 'My Line Plot')

# Dataset 2:
x_values_2 = [1, 2, 3, 4, 5]
y_values_2 = [5, 8, 12, 18, 25]
create_and_show_plot(ax2, x_values_2, y_values_2, 'New Line Plot', 'red', 'New Line Plot')

# Adjust layout for better spacing
plt.tight_layout()
```
